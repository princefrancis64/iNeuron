```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('wholesale_customers.txt',sep=',')
df.head()
```

```python
#Lets normalize the data since in case of distance based approach we use normalization
from sklearn.preprocessing import normalize
df1 = df
df1= normalize(df1)
```

```python
df1
```

```python
df2 = pd.DataFrame(df1,columns = df.columns)
df2.head()
```

```python
#building a model
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering()
model.fit(df2)
```

```python
model.labels_
```

```python
df2
```

```python
#Lets take up only 6 data points to have a better understanding
df3 = df2.iloc[:7,:]
df3
```

```python
#build a model
model_2 = AgglomerativeClustering()
model_2.fit(df3)
model_2.labels_
```

```python
#building a dendrogram
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df3, method='ward'))
```

```python
#linkage matrix
import scipy.cluster.hierarchy as shc
linkage_matrix = shc.linkage(df3,method='ward')
```

```python
#building a dendrogram with threshold 0.6
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dendrogram = shc.dendrogram(linkage_matrix)
plt.axhline(y=0.6, color='r', linestyle='--')
plt.show()
```

- This means we should have basically 2 clusters
- In our model also the labels show as only 2 clusters
