```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
#independent and dependent features
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
y.value_counts()
```

```python
iris.target
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
#Model Building
from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#Post pruning
treemodel2 = DecisionTreeClassifier(max_depth=2)
treemodel2.fit(X_train,y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel2)
```

- Here in the left hand side always get the value as versicolor 
- In the right hand side the value would be verginica

```python
#prediction
y_pred = treemodel2.predict(X_test)
y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#Pre Pruning
parameter= {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2'],
    'ccp_alpha':[1,2,3,4,5,6,7]
}
```

```python
#GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
tree_model3 = DecisionTreeClassifier()
clf = GridSearchCV(estimator=tree_model3,param_grid=parameter,cv = 3,n_jobs=-1,scoring='accuracy')
clf.fit(X_train,y_train)
```

```python
#Grid Search best params
clf.best_params_
```

```python
#model with best params
clf.predict(X_test)
```

- This looks like an overfitted data

```python
#selecting second set of params
parameter_2= {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['sqrt','log2'],
  
}
```

```python
clf_2 = GridSearchCV(estimator=tree_model3,param_grid=parameter_2,cv = 5,n_jobs=-1,scoring='accuracy')
clf_2.fit(X_train,y_train)
```

```python
clf_2.predict(X_test)
```

```python
y_pred_final = clf_2.predict(X_test)
```

```python
#accuracy score
score_final = accuracy_score(y_test,y_pred_final)
score_final
```

```python
print(classification_report(y_test,y_pred_final))
```

- It is basically the same accuracy as before
