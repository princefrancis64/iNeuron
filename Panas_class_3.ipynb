```python
def profit_flag(a):
    if a>0:
        return 'positive'
    elif a<0:
        return 'negative'
    else:
        return 'no profit'
```

```python
profit_flag(45)
```

```python
profit_flag(-78)
```

```python
df_sales['flag_profit'] = df_sales.profit.apply(profit_flag)  #apply function
```

```python
def len_customer(a):
    return len(a)
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len_customer) #df_sales['len_cust_name'] = df_sales['customer_name'].apply(len) 
```

```python
def quantity_flag(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:
        return 'medium'
    elif a>=20:
        return 'high'
```
